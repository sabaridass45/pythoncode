Use Case – Loan Approval:

   a)    Create a data frame  with  all Male customers, who are not graduates , whose loan is approved  
           and also get the Annual income and age of those customers
 
   b)    Final  Index  should be on Loan_ID
 
   c)    Impute missing values with appropriate measure use forward fill/backward fill/Generic
 
   d)    Let’s calculate the average spending score by Gender
 
   e)    Let’s calculate the average and total Loan amount by Property_Area .
 
   d)   Load back this data frame into excel with name Processed Loan
 

```python
import pandas as pd
import os
import glob
import numpy as np
```

```python
'''
a) Create a data frame with all Male customers, who are not graduates , whose loan is approved
'''

cust_df=pd.read_csv('D:\Sabari\Study_Material\Python\python_hcl\pandas\Pandas_session_20200925\Customers.csv')
loan_df=pd.read_csv('D:\Sabari\Study_Material\Python\python_hcl\pandas\Pandas_session_20200925\Loans.csv')

ln_aprvd_df = loan_df[(loan_df.Gender == 'Male') & (loan_df.Education == 'Not Graduate') & (loan_df.Loan_Status == 'Y')]
```

```python
'''
To get the Annual income and age of those customers using ln_aprvd_df and cust_filt_df via merge()
'''

inner_join_df = pd.merge(ln_aprvd_df,cust_df[['CustomerID' ,'Genre','Age', 'Annual Income (k$)','Spending Score (1-100)']], how='inner', left_on=['ID','Gender'], right_on=['CustomerID','Genre']).drop(['CustomerID','Genre'], axis = 1)
left_join_df = pd.merge(ln_aprvd_df,cust_df[['CustomerID' ,'Genre','Age', 'Annual Income (k$)','Spending Score (1-100)']], how='left', left_on=['ID','Gender'], right_on=['CustomerID','Genre']).drop(['CustomerID','Genre'], axis = 1)
```

```python
#print left_join_df or inner_join_df
left_join_df
inner_join_df
```

```python
#b) Final Index should be on Loan_ID

left_join_set_indx = left_join_df.set_index(['Loan_ID'])
inner_join_set_indx = inner_join_df.set_index(['Loan_ID'])
```

```python
#c.)Impute missing values with appropriate measure use forward fill/backward fill/Generic
left_join_ffil = left_join_set_indx.fillna( method ="ffill")
left_join_bfil = left_join_set_indx.fillna( method ="bfill")
left_join_filna = left_join_set_indx.fillna(0)
left_join_dropna = left_join_set_indx.dropna( how='any')
```

```python
#print either left_join_ffil or left_join_bfil or left_join_filna or left_join_dropna

#left_join_ffil
left_join_bfil
left_join_filna
#left_join_dropna
```

```python
#d) Let’s calculate the average spending score by Gender
left_join_filna.groupby("Gender").mean()
left_join_filna.groupby("Gender").agg({"Spending Score (1-100)":['mean']})
```

```python
#e) Let’s calculate the average and total Loan amount by Property_Area
left_join_filna.groupby("Property_Area")['LoanAmount'].agg([np.sum, np.mean])
```
